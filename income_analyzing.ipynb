```python
qazanc = [14574.49, 7606.46, 8611.41, 9175.41, 8058.65, 8105.44, 11496.28, 9766.09, 10305.32, 14379.96, 10713.97, 15433.50]
xərc = [12051.82, 5695.07, 12319.20, 12089.72, 8658.57, 840.20, 3285.73, 5821.12, 6976.93, 16618.61, 10054.37, 3803.96]
months = ['January','February','March','April','May','June','July','August','September','October','November','December']
```

```python
import numpy as np
revenue = np.array(qazanc)
expenses = np.array(xərc)
```

```python
#profit for each month
monthly_profit = revenue - expenses
monthly_profit
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = 12,6
plt.plot(months,monthly_profit,color = 'brown',marker = '.', markersize = 12)
plt.xticks(range(len(monthly_profit)))
plt.xlabel('Months')
plt.ylabel('Profit')
plt.grid()
plt.show()
```

```python
#profit after tax for each month (the tax rate is 30%)
profit_after_tax = monthly_profit*0.70
profit_after_tax[0]
```

```python
plt.rcParams['figure.figsize'] = 12,6
plt.plot(months,profit_after_tax,color = 'red',marker = '.', markersize = 12)
plt.xticks(range(len(profit_after_tax)))
plt.xlabel('Months')
plt.ylabel('Profit after tax')
plt.grid()
plt.show()
```

```python
#profit margin for each month 
profit_margin = profit_after_tax/revenue*100
profit_margin
```

```python
plt.rcParams['figure.figsize'] = 12,6
plt.plot(months,profit_margin,color = '#0FAEC6', marker = '.', markersize = 12)
plt.xticks(range(len(profit_margin)))
plt.yticks(range(-31,63,10))
plt.xlabel('Months')
plt.ylabel('Profit margin %')
plt.grid()
plt.show()
```

```python
#good months
average = profit_after_tax.mean()
average
```

```python
good_months_names = []
good_months = []
for i in range(len(profit_after_tax)):
  if profit_after_tax[i]>average:
    good_months_names.append(months[i])
    good_months.append(profit_after_tax[i])
print('Good months: ',dict(zip(good_months_names,good_months)))
```

```python
#bad months
bad_months_names = []
bad_months = []
for i in range(len(profit_after_tax)):
  if profit_after_tax[i]<average:
    bad_months_names.append(months[i])
    bad_months.append(profit_after_tax[i])
print('Bad months: ',dict(zip(bad_months_names,bad_months)))
```

```python
#the best month, the worst month
pat = dict(zip(list(profit_after_tax),months))
the_best_month = max(profit_after_tax)
the_worst_month = min(profit_after_tax)
print('The best month: ',pat[the_best_month],'-->',the_best_month)
print('The worst month: ',pat[the_worst_month],'-->',the_worst_month)
```

```python
#bonus - add 10% of profit (after tax) mean for good months
bonus = average*0.10
good_months_after_bonus = np.array(good_months)+bonus
good_months_after_bonus
```

```python
#penalty - subtract 10% of profit (after tax) mean for good months
penalty = average*0.10
bad_months_after_penalty = np.array(bad_months) - penalty
bad_months_after_penalty
```
